```
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(42)
```

## Przygotowanie danych

```
X1 = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)

print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba próbek: {m}')
```

```
# zmiana kształtu X1
X1 = X1.reshape(m, 1)
Y = Y.reshape(-1, 1)
print(X1)
print(X1.shape)
```

```
# Dodanie wyrazów wolnych
bias = np.ones((m, 1))
print(bias)
print(bias.shape)
```

```
# Połaczneie macierzy
X = np.append(bias, X1, axis=1)
print(X)
print(X.shape)
```

##Losowa inicjalizacja parametrów

```
eta = 0.01 # co jaki x będziemy szukać minimum

weights = np.random.randn(2,1)
print(X)
print(weights)
```

```
gradient = (2 / m) * X.T.dot(X.dot(weights)-Y)
gradient
```

##Metoda gradientu prostego

```
intercept = []
coef = []

for i in range(3000):
  gradient = (2 / m) * X.T.dot(X.dot(weights)-Y) # obliczenie gradientu
  weights = weights - eta * gradient # aktualziacja wag
  intercept.append(weights[0][0])
  coef.append(weights[1][0])

print(weights)
```

```
df = pd.DataFrame(data={'intercept':intercept, 'coef':coef})
df.head()
```

```
px.line(df, y='intercept')
```

```
px.line(df, y='coef')
```
